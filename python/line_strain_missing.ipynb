```python
import pandas as pd
import numpy as np
import datajoint as dj
from ibl_pipeline import subject, acquisition, behavior
from ibl_pipeline.analyses import behavior as behavior_analysis
```

```python
sjs = subject.SubjectUser * subject.SubjectLab() * subject.Subject & 'subject_line is NULL or subject_strain is NULL' & 'subject_nickname not LIKE "human%"'
```

```python
sjs = sjs.proj('lab_name', 'responsible_user', 'subject_nickname', 'subject_line', 'subject_strain', 'subject_source')
sjs = sjs.fetch(format='frame').reset_index()
sjs = sjs.sort_values(by=['lab_name', 'responsible_user', 'subject_nickname'])
```
